Merging Tables & Initial Exploration 


```python
movies = pd.read_csv("../Data/tmdb_5000_movies.csv")
credits = pd.read_csv("../Data/tmdb_5000_credits.csv")
df=movies.merge(credits,left_on="id",right_on="movie_id")
```

Describing Data & Handling Inconsistent Values

```python
df.describe()
```

```python
cols = ["budget", "popularity", "revenue", "runtime", "vote_average", "vote_count"]
df_zero = df[(df[cols] == 0).any(axis=1)]
df_zero = df_zero[["original_title"] + cols]
df_zero
```

```python
cols = ["budget", "popularity", "revenue", "runtime", "vote_average", "vote_count"]

for col in cols:
    median_value = df[col].median()
    df.loc[df[col] == 0, col] = median_value
```

Data Cleaning

```python
df.drop(["title_y","movie_id"],axis=1,inplace=True,errors="ignore")
df.columns=df.columns.str.strip()
pd.set_option("display.max_columns", None)
df.drop(columns=["homepage"], inplace=True ,errors="ignore")
df.dropna(subset=["release_date"])
```

Loading Data from JSON to Python

```python
json_cols = ["genres", "keywords", "production_companies", "spoken_languages", "cast", "crew","production_countries"]
for col in json_cols:
    df[col]=df[col].apply(lambda x:ast.literal_eval(x) if isinstance(x,str) else x)
```

```python
df["Genres_names"]=df["genres"].apply(lambda x:[d["name"]for d in x])
```

```python
df["keywords"] = df["keywords"].apply(lambda x: [d["name"] for d in x])
```

```python
df["production_Countries"]=df["production_countries"].apply(lambda x:[d["name"]for d in x])
```

```python
df["Main_actor"]=df["cast"].apply(lambda x: x[0]["name"] if len(x)>0 else None)
```

```python
df["All_cast"] = df["cast"].apply(lambda x:[d["name"]for d in x[:]])
```

```python
df["Director"]=df["crew"].apply(lambda x:next((d["name"]for d in x if d["job"]=="Director"),None))
```

```python
df["languages"]=df["spoken_languages"].apply(lambda x :[d["name"]for d in x])
```

```python
df["main_production_company"] = df["production_companies"].apply(lambda x:[d["name"]for d in x])
```

Feature Engineering


```python
df["release_date"] =pd.to_datetime(df["release_date"],errors="coerce")
df["release_year"] = df["release_date"].dt.year
df["release_month"] = df["release_date"].dt.month

```

```python
df["release_year"] = df["release_year"].astype("Int64")
df["release_month"] = df["release_month"].astype("Int64")
```

```python
df.isna().sum()
```

Filling Na

```python
df["overview"]=df["overview"].fillna("No overview available")
df["Main_actor"] = df["Main_actor"].fillna("Unknown Actor")
df["Director"] = df["Director"].fillna("Unknown Director")
df["tagline"] = df["tagline"].fillna("")
df["budget"] = df["budget"].fillna(0)
df["revenue"] = df["revenue"].fillna(0)
df.to_csv("../Data/movies_clean.csv", index=False)
```

Profit Insights:

```python
# Top10 movies by profit

df["profit"] = df["revenue"] - df["budget"]

top10_movies_profit = df.sort_values("profit", ascending=False).head(10)



# Top10 genres by profit

df_exploded = df.explode("Genres_names")
avg_profit_genre = (
    df_exploded.groupby("Genres_names")["profit"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)
# Runtime Stats
runtime_stats = df[df["runtime"].notna()].copy()
runtime_stats["runtime_category"] = pd.cut(
    runtime_stats["runtime"],
    bins=[0, 100, 140, 300],
    labels=["Short", "Medium", "Long"]
)

# Subplots 2×2
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Top 10 Movies by Profit", "Top 10 Genres by Average Profit","Runtime vs Profit","Profit vs Vote Average")
)
# Subplot 1: Top10 movies by profit
fig.add_trace(
    go.Bar(x=top10_movies_profit["title_x"], y=top10_movies_profit["profit"], marker_color="teal"),
    row=1, col=1
)

# Subplot 2: Top10 genres by profit

fig.add_trace(
    go.Bar(x=avg_profit_genre.index, y=avg_profit_genre.values, marker_color="steelblue"),
    row=1, col=2
)

# Subplot 3: Runtime vs Profit 
fig.add_trace(
    go.Box(
        x=runtime_stats["runtime_category"],
        y=runtime_stats["profit"],
        marker_color="royalblue",
        name="Profit Distribution"
    ),
    row=2, col=1
)



# Subplot 4: Profit vs Vote Average
fig.add_trace(
    go.Scatter(
        x=runtime_stats["vote_average"],
        y=runtime_stats["profit"],
        mode="markers",
        marker=dict(color="purple"),
        name="Profit vs Vote Average"
    ),
    row=2, col=2
)



fig.update_layout(
    height=1000, width=1000,
    title_text="Profit Insights",
    font=dict(size=12, color="white"),
    margin=dict(t=80, b=80, l=60, r=60),
    template="plotly_dark"
)

fig.update_xaxes(tickangle=-30)

fig.show()
```

```python

financial_df = ["popularity","budget","revenue","profit","vote_count","runtime"]
corr_matrix = df[financial_df].corr()

# 1) Top 10 Movies by Revenue
top10_revenue = df.sort_values("revenue", ascending=False).head(10)

# 2) Top 10 Genres by Profit
df["Genres_combo"] = df["Genres_names"].apply(lambda x: " | ".join(x) if isinstance(x, list) else x)
top10_genres_profit = (
    df.groupby("Genres_combo")["profit"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

# 3) Average Revenue & Budget by Genre
df_exploded = df.explode("Genres_names")
avg_rev_budget_genre = (
    df_exploded.groupby("Genres_names")[["revenue", "budget"]]
    .mean()
    .sort_values("revenue", ascending=False)
    .head(10)
)

# Subplots 2×2
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        "Top 10 Movies by Revenue",
        "Top 10 Genres combination by Average Profit",
        "Top 10 Genres by Average Revenue & Budget",
        "Financial Metrics Correlation Heatmap"
    )
)

# Subplot 1: Top 10 Movies by Revenue (Revenue + Budget)
fig.add_trace(
    go.Bar(x=top10_revenue["title_x"], y=top10_revenue["revenue"], name="Revenue", marker_color="seagreen"),
    row=1, col=1
)
fig.add_trace(
    go.Bar(x=top10_revenue["title_x"], y=top10_revenue["budget"], name="Budget", marker_color="orange"),
    row=1, col=1
)

# Subplot 2: Top 10 Genres by Profit
fig.add_trace(
    go.Bar(x=top10_genres_profit.index, y=top10_genres_profit.values, name="Profit", marker_color="mediumseagreen"),
    row=1, col=2
)

# Subplot 3: Average Revenue & Budget by Genre
fig.add_trace(
    go.Bar(x=avg_rev_budget_genre.index, y=avg_rev_budget_genre["revenue"], name="Revenue", marker_color="dodgerblue"),
    row=2, col=1
)
fig.add_trace(
    go.Bar(x=avg_rev_budget_genre.index, y=avg_rev_budget_genre["budget"], name="Budget", marker_color="orange"),
    row=2, col=1
)

# Subplot 4: Correlation Heatmap
fig.add_trace(
    go.Heatmap(
        z=corr_matrix.values,
        x=corr_matrix.columns,
        y=corr_matrix.columns,
        colorscale="RdBu",
        zmin=-1, zmax=1,
        text=corr_matrix.round(2).values,
        texttemplate="%{text}",
        hoverongaps=False
    ),
    row=2, col=2
)

fig.update_layout(
    height=1000, width=1000,
    title_text="Revenue, Profit & Budget Insights",
    font=dict(size=12, color="white"),
    margin=dict(t=80, b=80, l=60, r=60),
    barmode="group", template="plotly_dark"
)
fig.update_xaxes(tickangle=-30)

fig.show()
```

Conclusion & Recommendations



🎬 Insights from Correlation Matrix of Financial Metrics
The correlation matrix reveals key relationships between core financial and popularity metrics in movie data:
- Revenue vs Profit (0.97): Extremely strong positive correlation. Profit is almost directly tied to revenue, which is expected since profit = revenue − budget.
- Budget vs Revenue (0.73): Strong correlation. Higher-budget films tend to generate higher revenues, though not always proportionally.
- Popularity vs Vote Count (0.77): Strong correlation. Popular films attract more audience engagement and votes.
- Popularity vs Revenue (0.64): Moderate-to-strong correlation. Popularity contributes to revenue, but other factors (e.g., quality, marketing) also play a role.
- Budget vs Vote Count (0.59): Moderate correlation. Larger budgets often lead to wider exposure and more votes, but not as strongly as popularity does.

  Conclusion:
Financial success (revenue and profit) is closely linked to budget, while audience engagement (vote count) is strongly influenced by popularity. Popularity also affects revenue, but it's not the sole driver.


🎬 Insights from "Top 10 Genre Combinations by Average Profit"
The genre combination chart highlights which blends of genres yield the highest average profits:
- Genre combinations outperform single genres. All top entries are multi-genre blends, showing that mixing genres enhances audience appeal and profitability.
- Family-oriented and fantasy genres dominate. Combinations like Family | Fantasy | Adventure and Animation | Comedy | Family consistently rank high, indicating strong market demand for emotionally engaging, family-friendly content.
- Blending action, drama, and fantasy is lucrative. Combinations like Action | Drama | Fantasy and Action | Adventure | Fantasy | Science Fiction show that mixing intensity with imagination drives profitability.
- Science Fiction adds value when paired well. It appears in multiple top combinations, suggesting it boosts profit when combined with adventure or fantasy.
- Avoid genre over-specialization. No single-genre combinations appear in the top 10, implying that narrow genre focus may limit commercial success.

Conclusion:
Genre blending is a key driver of profitability. Investors and producers should prioritize multi-genre projects—especially those combining family, fantasy, adventure, and science fiction—to maximize return on investment.
